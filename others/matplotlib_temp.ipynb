### `show()` or `No show()`? How to Display Your Plots

Việc sử dụng Matplotlib tốt nhất khác nhau tùy thuộc vào cách bạn đang sử dụng nó; đại khái là ba ngữ cảnh áp dụng đang sử dụng Matplotlib trong một `script`, trong một `IPython terminal` hoặc trong một `IPython notebook`.

**Plotting from a script**

```python
# ------- file: myplot.py ------
    import matplotlib.pyplot as plt
    import numpy as np
    x = np.linspace(0, 10, 100)
    plt.plot(x, np.sin(x))
    plt.plot(x, np.cos(x))
    plt.show()
```

## Two Interfaces for the Price of One

### MATLAB-style interface

Matplotlib ban đầu được viết như một sự thay thế Python cho người dùng MATLAB và phần lớn cú pháp của nó phản ánh thực tế đó. Các công cụ kiểu MATLAB được chứa trong giao diện pyplot (plt). Ví dụ, đoạn mã sau có thể trông khá quen thuộc với người dùng MATLAB

````python
plt.figure() # create a plot figure
# create the first of two panels and set current axis 
plt.subplot(2, 1, 1) # (rows, columns, panel number) 
plt.plot(x, np.sin(x))
# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
````

### Object-oriented interface

`Object-oriented interface` có sẵn cho những tình huống phức tạp hơn này và khi bạn muốn kiểm soát nhiều hơn hình của mình. Thay vì phụ thuộc vào một số khái niệm về một hình hoặc các trục “active”, trong `object-oriented interface`, các hàm vẽ đồ thị là các phương thức của các đối tượng `Figure` và `Axes` rõ ràng.

````python
fig, ax = plt.subplots(2) # create a figure and two axes
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
````

### `plot` vs `scatter`: A Note on Efficiency

- Ngoài các tính năng khác nhau có sẵn trong `plt.plot` và `plt.scatter`, tại sao bạn có thể chọn sử dụng cái này thay cho cái kia? Mặc dù điều đó không quan trọng lắm đối với lượng dữ liệu nhỏ, vì tập dữ liệu lớn hơn vài nghìn điểm, `plt.plot` có thể hiệu quả hơn đáng kể so với `plt.scatter`. 

Lý do là:
- `plt.scatter` có khả năng hiển thị kích thước và / hoặc màu sắc khác nhau cho mỗi điểm, vì vậy trình kết xuất phải thực hiện thêm công việc xây dựng từng điểm riêng lẻ. 
- Mặt khác, trong `plt.plot`, các điểm về cơ bản luôn là bản sao của nhau, vì vậy công việc xác định sự xuất hiện của các điểm chỉ được thực hiện một lần cho toàn bộ tập dữ liệu. Đối với các tập dữ liệu lớn, sự khác biệt giữa hai thứ này có thể dẫn đến hiệu suất rất khác nhau và vì lý do này, `plt.plot` nên được ưu tiên hơn `plt.scatter` cho các tập dữ liệu lớn.

## Visualizing Errors

Đối với bất kỳ phép đo khoa học nào, việc tính toán chính xác các sai sót gần như quan trọng, nếu không muốn nói là quan trọng hơn việc báo cáo chính xác con số. Ví dụ: hãy tưởng tượng rằng tôi đang sử dụng một số quan sát vật lý thiên văn để ước tính Hằng số Hubble, phép đo cục bộ về tốc độ giãn nở của vũ trụ. Tôi biết rằng tài liệu hiện tại đề xuất giá trị khoảng 71 (km / s) / Mpc và tôi đo giá trị 74 (km / s) / Mpc bằng phương pháp của mình. Các giá trị có nhất quán không? Câu trả lời đúng duy nhất, với thông tin này, là: không có cách nào để biết

Giả sử tôi bổ sung thông tin này với các độ không đảm bảo được báo cáo: thông tin hiện tại cho thấy giá trị khoảng 71 ± 2,5 (km / s) / Mpc và phương pháp của tôi đã đo được giá trị là 74 ± 5 (km / s) / Mpc. Bây giờ các giá trị có nhất quán không? Đó là một câu hỏi có thể được trả lời một cách định lượng.
Trong việc trực quan hóa dữ liệu và kết quả, việc hiển thị những lỗi này một cách hiệu quả có thể làm cho một âm mưu truyền tải thông tin đầy đủ hơn nhiều.

### Basic Errorbars

````python
x = np.linspace(0, 10, 50)
dy = 0.8 
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k') # error bar, yerr: error, fmt: format
````

````python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black', ecolor='lightgray', elinewidth=2, capsize=0) 
# error bar, capsize: cap size
````

### Continuous Errors

Trong một số tình huống, mong muốn hiển thị các thanh báo lỗi về số lượng liên tục. Mặc dù Matplotlib không có quy trình tiện lợi tích hợp sẵn cho loại ứng dụng này, nhưng việc kết hợp các nguyên mẫu như plt.plot và plt.fill_between để có kết quả hữu ích là tương đối dễ dàng.
Ở đây, chúng tôi sẽ thực hiện hồi quy quy trình Gaussian đơn giản (GPR), sử dụng API Scikit-Learn (xem “Giới thiệu Scikit-Learn” trên trang 343 để biết thêm chi tiết). Đây là một phương pháp kết hợp một hàm phi tham số rất linh hoạt với dữ liệu có độ không đảm bảo đo liên tục. Chúng tôi sẽ không đi sâu vào chi tiết của hồi quy quy trình Gaussian tại thời điểm này, nhưng thay vào đó sẽ tập trung vào cách bạn có thể hình dung một phép đo lỗi liên tục như vậy:

Example (Do đây là ví dụ chạy trên sklearn phiên bản cũ, sẽ update theo phiên bản mới sau):
```python
from sklearn.gaussian_process import GaussianProcess
# define the model and draw some data

model = lambda x: x * np.sin(x)
xData = np.array([1,3,5,6,8])
yData = model(xData)

#Computer the Gausian Process fit
gp = GaussianProcess(corr='cubic', theta0=1e-2, thetaL=1e-4, thetaU=1E-1,random_start=100)
gp.fit(xData[:, np.newaxis], yData)

xfit = np.linspace(0, 10, 1000)
yfit, MSE = gp.predict(xfit[:, np.newaxis], eval_MSE=True)
dyfit = 2 * np.sqrt(MSE)

# Visualize the result plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color='gray')
plt.fill_between(xfit, yfit - dyfit, yfit + dyfit,
                            color='gray', alpha=0.2)
plt.xlim(0, 10)
```

## Density (Mật độ) and Contour (Đường viền) Plots

- Đôi khi, rất hữu ích khi hiển thị dữ liệu `ba chiều` trong `hai chiều` bằng cách sử dụng các contours or color-coded regions. 
- Có ba hàm `Matplotlib` có thể hữu ích cho nhiệm vụ này: `plt.contour` cho các ô đường đồng mức, `plt.contourf` cho các ô đường đồng mức đã điền và `plt.imshow` để hiển thị hình ảnh. 
- Phần này xem xét một số ví dụ về việc sử dụng chúng. Chúng tôi sẽ bắt đầu bằng cách thiết lập sổ ghi chép để vẽ và nhập các chức năng chúng tôi sẽ sử dụng:


### Visualizing a Three-Dimensional Function

$z = f(x, y)$

````python
def f(x,y):
    return np.sin(x)**10 + np.cos(10 + y * x) * np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y  = np.meshgrid(x, y) # two-dimensional grids from one-dimensional arrays:
X.shape
````

````python
Y.shape
````

````python
Z = f(X, Y)
````

````python
plt.contour(X, Y, Z, colors='black') # contour plot
````

Lưu ý rằng theo mặc định khi sử dụng một màu, các giá trị âm được biểu thị bằng các đường đứt nét và các giá trị dương bằng các đường liền. Ngoài ra, bạn có thể mã màu cho các dòng bằng cách chỉ định một bản đồ màu với đối số là cmap. Ở đây, chúng tôi cũng sẽ chỉ định rằng chúng tôi muốn vẽ nhiều dòng hơn — 20 khoảng cách đều nhau trong phạm vi dữ liệu

````python
plt.contour(X, Y, Z, 20, cmap='RdGy') # 20 contour lines``, RdGy short for Red-Gray
````

Plot của chúng tôi trông đẹp hơn, nhưng khoảng cách giữa các dòng có thể hơi mất tập trung. Chúng ta có thể thay đổi điều này bằng cách chuyển sang một biểu đồ đường bao đã được lấp đầy bằng cách sử dụng hàm `plt.contourf()` (lưu ý f ở cuối), sử dụng phần lớn cú pháp giống như `plt.contour()`.

````python
plt.contourf(X, Y, Z, 20, cmap='RdGy') # 20 contour lines, filled
plt.colorbar()
````

Ngoài ra, chúng tôi sẽ thêm lệnh `plt.colorbar()`, lệnh này sẽ tự động tạo một trục bổ sung với thông tin màu được gắn nhãn cho plot

````python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
                      cmap='RdGy', alpha=0.5)
plt.colorbar()
````
